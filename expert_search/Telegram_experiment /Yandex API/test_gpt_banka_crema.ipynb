```python
folder_id = 'b1gldmeh8cjs8aec6u88'
yandexgpt_key = 'AQVN1WgHsPi56JTt4Xmnq4qPIbURG4EKpvtFkLVg'
yandex_gpt_api_url = 'https://llm.api.cloud.yandex.net/foundationModels/v1/fewShotTextClassification'



def yandex_gpt(x):
    response = requests.post(
        yandex_gpt_api_url,
        headers={
            "Authorization": f"Api-Key {yandexgpt_key}",
            "x-folder-id": folder_id
        },
        json={
            "modelUri": f"cls://{folder_id}/yandexgpt/latest",
 
 "taskDescription": "Классифицируй блогера на основе написанных им постов.Профессиональные ниши Бьюти-тематики: косметолог, визажист, косметик-эстетист, ламимейкер, массажист, подолог.Критерии, по которым вы определяете экспертность : Эксперт имеет опыт, имеет навык доносить информацию до других людей.Он описывает методики, дает отзыв, рекомендация как использовать продукт",
  "labels": [
    "эксперт", "не эксперт"
  ],
  "text": f'{x}',
     
    }
     )
    return response.json()
```

```python
test1 = pd.read_csv('Обзоры на товары (Ответы) - Ответы на форму (1).csv')
```

```python
test1 = test1.drop('Отметка времени', axis=1 )
```

```python
answer = []
for i in range(len(test1)):
    answer.append(yandex_gpt(test1.loc[i]))
    time.sleep(20)
```

```python
pred_df = pd.DataFrame(answer)
```

```python
pred_df = pred_df.drop('modelVersion', axis=1)
```

```python
predict = pd.json_normalize(pred_df['predictions'])
```

```python
expert_pred = pd.json_normalize(predict[0])
```

```python
result = pd.concat([test1,expert_pred], axis=1)
```

## –û—Ç—Ñ–∏–ª—å—Ç—Ä–æ–≤–∞–Ω–Ω—ã–µ –ø–æ—Å—Ç—ã

```python
morpher = MorphAnalyzer()
```

```python
p = morpher.parse('—Å–æ–≤–µ—Ç—É—é')[0]
print(p.normal_form)
p1 = morpher.parse('–∫—É—Ä—Å–æ–≤')[0]
print(p1.normal_form)
p2 = morpher.parse('–º–µ–Ω—è –∑–æ–≤—É—Ç')[0]
print(p2.normal_form)
p3 = morpher.parse('–º–æ–µ –∏–º—è')[0]
print(p3.normal_form)
p4 = morpher.parse('–æ–ø—ã—Ç —Ä–∞–±–æ—Ç—ã')[1]
print(p4.normal_form)
```

```python
def preproccessing(df):
    text = df.lower()
    text = (
        ' '.join([
            morpher.parse(word)[0].normal_form 
            for word in text.split()]))
    return text
    
    
```

```python
w = {'—ç–∫—Å–ø–µ—Ä—Ç': ['—ç–∫—Å–ø–µ—Ä—Ç', '–ø—Ä–µ–ø–æ–¥–∞–≤–∞—Ç–µ–ª—å', '–∞–≤—Ç–æ—Ä –∫—É—Ä—Å' , '—Å—É–¥—å—è —á–µ–º–ø–∏–æ–Ω–∞—Ç','—Ä–µ–∫–æ–º–µ–Ω–¥–æ–≤–∞—Ç—å', '—Å–æ–≤–µ—Ç–æ–≤–∞—Ç—å', '–æ–ø—ã—Ç —Ä–∞–±–æ—Ç–∞','—è –∑–≤–∞—Ç—å','–º–æ–µ –∏–º—è', '–º–µ–Ω—è –∑–≤–∞—Ç—å']}
```

```python
def expert_filter(text):
    for key, words in w.items():
        if any(word in text for word in words):
            return preproccessing(text)
```

```python
text_beauty_filter = df_beauty.apply(expert_filter).dropna()
text_beauty_filter
```

```python
text_podology_filter = df_podolog.apply(expert_filter).dropna()
#text_podology_filter.to_list()
```
